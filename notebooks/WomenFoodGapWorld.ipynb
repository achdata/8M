```python
# plot data into a sigle script
import pandas as pd
import plotly.express as px

# Load the data
file_path = "https://raw.githubusercontent.com/achdata/8M/main/data/FAOSTAT_data_es_2-27-2024.csv"
data = pd.read_csv(file_path)

# Filtering the data to include only the annual prevalence of moderate or severe and severe food insecurity
# for adult men and women globally and in each region.

# Defining the relevant strings to search for in the 'Producto' column
relevant_products = [
    'Prevalencia de la inseguridad alimentaria moderada o grave en la población adulta masculina (%) (valor anual)',
    'Prevalencia de la inseguridad alimentaria grave en la población adulta masculina (%) (valor anual)',
    'Prevalencia de la inseguridad alimentaria moderada o grave en la población adulta femenina (%) (valor anual)',
    'Prevalencia de la inseguridad alimentaria grave en la población adulta femenina (%) (valor anual)'
]

# Filtering the data
filtered_data = data[data['Producto'].isin(relevant_products)]

# Creating a pivot table to structure the data for plotting
pivot_data = filtered_data.pivot_table(
    values='Valor',
    index=['Año', 'Área'],
    columns='Producto',
    aggfunc='first'
).reset_index()



# Renaming the columns for easier access
pivot_data.columns = [
    'Year', 'Region',
    'Hombres: Severa',
    'Hombres: Moderada o Severa',
    'Mujeres: Severa',
    'Mujeres: Moderada o Severa'
]

# Sorting the data by Year and Region to ensure the plot lines follow the correct order
pivot_data.sort_values(by=['Year', 'Region'], inplace=True)

# Change the name of America Seotentrionale y Europa to America Spt. y Europa
pivot_data['Region'] = pivot_data['Region'].replace('América Septentrional y Europa', 'América Spt. y Europa')

# Get the index of the "Mundo" region
mundo_index = pivot_data[pivot_data['Region'] == 'Mundo'].index[0]

# Move the "Mundo" row to the first position
pivot_data = pivot_data.iloc[[mundo_index] + list(range(mundo_index+1, len(pivot_data)))]

# Create an interactive graph using Plotly Express
fig = px.line(pivot_data, x='Year', y=['Hombres: Severa', 'Hombres: Moderada o Severa', 'Mujeres: Severa', 'Mujeres: Moderada o Severa'],
              color_discrete_map={
        'Hombres: Severa': 'darkgreen',
        'Hombres: Moderada o Severa': '#52AE32',
        'Mujeres: Severa': '#3E236E',
        'Mujeres: Moderada o Severa': '#9882BE'
    },
    title='Prevalencia de la inseguridad alimentaria en adultos por sexo y nivel de severidad',
    facet_col='Region',
    facet_col_wrap=5,
    height=600,
    width=1200,


)


# Update the layout to improve readability
fig.update_layout(
    template="plotly_white",
    #title_pad_t=-50,
    title=dict(text="<b>Prevalencia de la inseguridad alimentaria en adultos<br>por sexo y nivel de severidad, 2014-2022",
               font=dict(size=18, color="#707070")),
    title_x=0.075,
    title_y=0.96,
    font_family="Lato",
    title_font_family="Lato",
    xaxis_title= None,
    yaxis_title= None,
    xaxis_title_text=" ",
    yaxis_title_text="Población adulta",
    font_color="#707070",


    legend=dict(
        title=None, orientation="h", y=1.05, yanchor="bottom",
        x=0.3, xanchor="center", font_size=10, font_color="#707070"
    ),


    hoverlabel=dict(
        bgcolor="white",
        font_size=14,
        font_family="Lato"
    ),


)


# Customize the y and x-axis labels
fig.update_xaxes(title_text="", color="#707070")
fig.update_yaxes(tickprefix="%", showgrid=True, color="#707070")

# Remove "Region=" from the title of each chart
fig.for_each_annotation(lambda a: a.update(text=a.text.split("=")[1]))

# Add an annotation to the "América Latina y Caribe" chart
fig.add_annotation(
    x=0.59,
    y=28,
    xref="paper",
    yref="y",
    text="""<b>Esta brecha indica que las mujeres en la región están enfrentando<br>desafíos adicionales en comparación con los hombres cuando<br>se trata de acceder a alimentos suficientes y nutritivos.</b>""",
    showarrow=True,
    arrowhead=2,
    xanchor="left",
    font_size=11,
    font_family="Lato",
    font_color="#707070",
    ax=40,
    ay=-50
)


# Highlight the region with a shape
fig.add_shape(
    type='rect',
    label_padding=5,
    x0=1.1,
    y0=0,
    x1=2.2,
    y1=1,
    xref='x2 domain',
    yref='paper',
    fillcolor="LightGrey",
    opacity=0.2,
    #layer='below',
    line_color="black",
)
fig.update_traces(mode="markers+lines")
fig.update_traces(hovertemplate=
    'Año: %{x}<br>' +
    '<b>Porcentaje: %{y}</b><br>'
)


# add a vertical "target" line
fig.add_shape(
    type="line", line_color="#ee7203", line_width=3, opacity=1, line_dash="dash",
    x0=0.58, x1=0.58, xref="paper", y0=11.2, y1=41.8, yref="y"
)


# add data source
fig.add_annotation(
    x=0,
    y=-0.15,
    xref="paper",
    yref="paper",
    text="Elaboración propio de Acción contra el Hambre - Fuente: <a href='https://www.fao.org/faostat/es/#data/FS'>FAOSTAT, 2024.</a> - Encuentra el script: <a href='https://github.com/achdata/8M'>enlace al código.</a>",
    showarrow=False,
    align="left",
    font_size=12,
    font_family="Lato",
    font_color="#707070",
    ax=0,
    ay=0
)

#https://github.com/achdata/8M

config = dict({'displayModeBar': False})

# Show the plot
fig.show(config=config)

# salva el grafico como html

fig.write_html("/content/inseguridad_alimentaria_adultos.html", config=config)

```
